```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import image
```

## SVD用于压缩数据
导入一张图片数据，分辨率为1080*1920。

```python
raw_img = image.imread('../datasets/img/test.jpg')
img_trans = raw_img.reshape(1920, 1080*3)    # 注意高在前，宽在后
U, Sigma, VT = np.linalg.svd(img_trans)
```

分别取前100个奇异值与前200个奇异值，利用三个矩阵的部分信息来重构图片。

```python
res_img_list = list()
for n_sig_val in [100, 200]:
    res_img = np.dot(np.dot(U[:, :n_sig_val], np.diag(
        Sigma[:n_sig_val])), VT[:n_sig_val, :]).reshape(1920, 1080, 3)

    res_img = res_img.astype(np.int16)    # 浮点转整形
    res_img[res_img < 0] = 0
    res_img[res_img > 255] = 255

    res_img_list.append(res_img)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 20))

axs[0].imshow(raw_img)
axs[1].imshow(res_img_list[0])
axs[2].imshow(res_img_list[1])

plt.show()
```

## SVD用于降维
使用SVD降维时，同PCA，右乘右奇异矩阵的前$k$列即可。

```python
from sklearn.datasets import make_circles
X, Y = make_circles(factor=0.5, random_state=0, noise=0.05)
U, Sigma, VT = np.linalg.svd(X)

Sigma=np.diag(Sigma)
X_trans=np.dot(X,VT[:,:2])

print(X_trans.shape)
```
